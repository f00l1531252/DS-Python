```python
def sigmoid(x):
    return  1 / (1 + np.e**-x)  
x = np.arange(-10, 10) 
plt.plot(x, sigmoid(x))
```

## 図の装飾

```python
x = np.linspace(-10, 10, 30)
z = 1 / (1 + np.exp(-x))
plt.plot(x, z, 
         color='r',
         linestyle='--',
         linewidth=3,
         marker='o',
         markersize=7,
         markerfacecolor='y',
         markeredgecolor='b',
         markeredgewidth=1,
         alpha=0.5
         )
```

## 付属情報をつける

```python
plt.plot(x, z, label='sigmoid')
plt.plot(x, z*2, label='sigmoid*2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=4)
plt.xticks(np.linspace(-10, 10, 12), rotation=20)
plt.yticks(np.linspace(0, 2, 10), rotation=20)
plt.show()
```

```python
df_movies = pd.read_csv('../dataset/tmdb/tmdb_5000_movies.csv')
df_credits = pd.read_csv('../dataset/tmdb/tmdb_5000_credits.csv')
df = df_movies.merge(df_credits, left_on='id', right_on='movie_id', suffixes=('_movies', '_credits'))
df.head(3)
```

```python
len(df)
```

```python
df = df[~((df['budget'] == 0) | (df['revenue'] == 0))]
len(df)
```

```python
df['profit'] = df.apply(lambda x: x['revenue'] - x['budget'], axis=1)
df.head(2)
```

```python
for idx, movie in df.iterrows():
    if movie['vote_average'] > 8:
        print(idx, movie['title_movies'])
```

```python
df.columns
```

```python
df['year'] = df.apply(lambda x: int(x['release_date'][:4]), axis=1) 
df.head(2)
```

```python
df[['year', 'release_date']].head(3)
```

```python
df.groupby('year').count()
```

```python
plt.plot(df.groupby('year').count()['id'], label='movies')
plt.xlabel('year')
plt.ylabel('count')
plt.title('movie_count')
plt.legend()
plt.show()
```

```python
plt.plot(df.groupby('year').median()['revenue'], label='median')
plt.plot(df.groupby('year').mean()['revenue'], label='mean')
plt.plot(df.groupby('year')['revenue'].max(), label='max')
plt.xlabel('year')
plt.ylabel('revenue')
plt.legend()
```

## 複数の図の描画する

### .subplot()

```python
x = np.linspace(-10, 10, 100)
plt.subplot(1, 2, 1) # .subplot(nrow, ncol, index)
plt.plot(x, x**2)

plt.subplot(1, 2, 2)
plt.plot(x, x**3)
```

### .figure()

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)


ax1.plot(x, x**2)
ax2.plot(x, x**3)
```

### .subplots()

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(x, x**2, label='x**2')
axes[0, 1].plot(x, x**3)
axes[1, 0].plot(x, x**4)
axes[1, 1].plot(x, x**5)

axes[0, 0].set_xlabel('xlabel1')
axes[0, 0].set_ylabel('ylabel1')

axes[0, 0].set_xticks([-10, 0, 10])
axes[0, 0].set_yticks(np.arange(0, 101, 20))

axes[0, 0].legend()


axes[0, 0].set_title('x**2')
axes[0, 1].set_title('x**3')
axes[1, 0].set_title('x**4')
axes[1, 1].set_title('x**5')

axes[1, 1].axis('off') # 画像を表示する時に使う

fig.tight_layout()
```

## 色々なplotを書く

### .scatter()：散布図

```python
df = pd.read_csv('../data/tmdb_5000_movies_profit_sorted.csv')
df.head(2)
```

```python
plt.scatter('revenue', 'budget', data=df, alpha=0.2)
plt.xlabel('revenue')
plt.ylabel('budget')
```

```python
language_group_df = df.groupby('original_language').mean()[['revenue', 'budget', 'vote_count']]
language_group_df.head()
```

```python
plt.scatter('revenue', 'budget', 
            s='vote_count', 
            c=np.arange(len(language_group_df)), 
            data=language_group_df, alpha=0.7)
for idx, row in language_group_df.iterrows():
    plt.annotate(idx, (row['revenue'], row['budget']))
```

```python
np.arange(len(language_group_df))
```

```python
for lang, row in language_group_df.iterrows():
    print(lang, row['revenue'], row['budget'])
```

## plotを保存する

### サイズを指定してplotする

```python
x = np.linspace(-3, 3, 10)
plt.figure(figsize=(10, 5), dpi=100) # dpi= dot per inch
plt.plot(x, x**2)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
```

### plt.savefig('path)：pngファイルで保存

```python
plt.scatter('runtime', 'budget', data=df, alpha=0.3)
plt.xlabel('runtime')
plt.ylabel('budget')
plt.savefig('../data/runtime_budget_scatter.png')
```

### fig.savefig('path')

```python
x = np.linspace(-3, 3, 100)
fig, axes = plt.subplots(2, 1, figsize=(10, 3))
axes[0].plot(x, np.sin(x))
axes[0].set_title('sin')
axes[1].plot(x, np.cos(x))
axes[1].set_title('cos')
fig.tight_layout()
fig.savefig('../data/sin_cos.png')
```
